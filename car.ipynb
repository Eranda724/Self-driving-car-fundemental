```python
import cv2
image = cv2.imread("car.jpg")
```

```python

print("Image type", type(image))
cv2.imshow("Image name", image)
cv2.waitKey(0)
```

```python
import matplotlib.pyplot as plt
%matplotlib inline

plt.imshow(image)
```

colour change because in opencv use BGR

```python
image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

plt.imshow(image_rgb)
```

```python
image_gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
print("image shape", image_gray.shape)
plt.imshow(image_gray, cmap="gray")
```

```python
sml_img = cv2.resize(image_rgb
                     ,(100, 100))

plt.imshow(sml_img)
```

```python
ksize = (15,15)

img_blr = cv2.blur(image_rgb, ksize)

flg, (ax1, ax2) = plt.subplots(2, 1, figsize = (15,10))
ax1.imshow(img_blr)

ax2.imshow(img_blr)
```

```python
img_edg = cv2.Canny(image_rgb, 125,175)
plt.imshow(img_edg)
```

```python
import cv2

start_point = (45, 50)

end_point = (265, 130)

color = (255, 0,0)

thickness = 1

image_rec = cv2.rectangle(image_rgb, start_point,end_point, color, thickness)

plt.imshow(image_rec)
```
